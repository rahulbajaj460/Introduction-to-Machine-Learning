## Decision tree implementation using sklearn datasets

```python
#Loading the dataset 
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data.DESCR
## It describes the dataset.
```

```python
data.data
## it contains the input variables
```

```python
data.feature_names
## it contains the column names
```

```python
data.target
## it is the target output
```

```python
data.target_names
## it is the target output categories
```

```python
## For loading the dataset as Pandas dataframe.
import pandas as pd
import numpy as np
df = pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
df.head()
```

```python
df.tail()
# displays the last five rows of the data.
```

```python
df.shape
## tells about the shape of the dataframe
```

```python
X=df.iloc[:,:-1]
## doesn't take the last column of the dataframe.
X.head()
```

```python
y = df.iloc[:,-1]
## takes the last column of the dataframe
y
```

```python
## Splitting the dataset.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print('Shape of X_train ',X_train.shape)
print('Shape of y_train ',y_train.shape)
print('Shape of X_test ',X_test.shape)
print('Shape of y_test ',y_test.shape)
```

## Decision tree using gini criterion

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

## Decision tree using entropy criterion

```python
model1=DecisionTreeClassifier(criterion='entropy')
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```
